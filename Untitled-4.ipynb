```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('nei.csv', sep = ';')

years = [1999, 2002, 2005, 2008]
data['Emissions'] = data['Emissions'].apply(lambda x: float(x.replace(',', '.')))
values = []
for year in years:
    data_year = data[data['year'] == year]
    values.append(sum(data_year['Emissions']))
figure = plt.figure(1, figsize=(20, 10))
plt.subplot(111)
plt.bar(years, values)
plt.suptitle('Result1')
plt.show()
figure.savefig('plot1.png', bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#data = pd.read_csv('nei.csv', sep = ';')
data.head()
years = [1999, 2002, 2005, 2008]
#data['Emissions'] = data['Emissions'].apply(lambda x: float(x.replace(',', '.')))

types = ['point', 'nonpoint', 'on-road', 'non-road']
# point is the first line
# nonpoint is the second line
# on-road is the third line
# non-road is the fourth line
data['type'] = data['type'].apply(lambda x: x.lower())
data['fips'] = data['fips'].apply(lambda x: str(x))
figure = plt.figure(1, figsize=(20, 10))
config = [421, 423, 425, 427] #left side for specific location
config_all = [422, 424, 426, 428] #right side for all
for ind in range(len(types)):
    values = []
    values_all = []
    for year in years:
        data_year = data[data['year'] == year]
        data_year_type = data_year[data_year['type'] == types[ind]]
        data_year_location = data_year[data_year['fips'] == "24510"]
        data_year_location_type = data_year_location[data_year_location['type'] == types[ind]]
        values.append(sum(data_year_location_type['Emissions']))
        values_all.append(sum(data_year_type['Emissions']))
    print values
    plt.subplot(config[ind])
    plt.bar(years, values)
    plt.subplot(config_all[ind])
    plt.bar(years, values_all)

plt.suptitle('Result3')
figure.savefig('plot3.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('nei.csv', sep = ';')
data['SCC'] = data['SCC'].apply(lambda x: str(x))
data['fips'] = data['fips'].apply(lambda x: str(x))
data['Emissions'] = data['Emissions'].apply(lambda x: float(x.replace(',', '.')))
data_info = pd.read_csv('scc.csv', sep = ';')
data_all = pd.merge(data, data_info, left_on = 'SCC', right_on = 'SCC', how = 'inner')
years = [1999, 2002, 2005, 2008]
values1 = []
values2 = []
for year in years:
    data_year = data_all[data_all['year'] == year]
    data_year_source =  data_year[data_year['SCC.Level.One'] == 'Mobile Sources']
    data_year_source_location1 = data_year_source[data_year_source['fips'] == "24510"]
    data_year_source_location2 = data_year_source[data_year_source['fips'] == "06037"]
    values1.append(sum(data_year_source_location1['Emissions']))
    values2.append(sum(data_year_source_location2['Emissions']))
figure = plt.figure(1, figsize=(20, 10))
plt.subplot(211)
plt.bar(years, values1)
plt.subplot(212)
plt.bar(years, values2)
plt.suptitle('Result5&6')
plt.show()
figure.savefig('plot5&6.png', bbox_inches='tight')
```
